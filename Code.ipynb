**GETTING INPUT FROM THE USER**

```python
# Function to check water safety
def check_water_safety():
    # Safe limits based on WHO and general standards
    safe_limits = {
        "pH": (6.5, 8.5),
        "Sulfate": (0, 500),
        "Hardness": (0, 300),
        "Solids": (0, 500),
        "Chloramines": (0, 4),
        "Conductivity": (0, 1000),
        "Organic Carbon": (0, 5),
        "Trihalomethanes": (0, 80),
        "Turbidity": (0, 5)
    }

    # Dictionary to store user inputs
    values = {}

    # Get input values from user
    print("\nEnter the water quality parameters:")

    for param in safe_limits.keys():
        while True:
            try:
                value = float(input(f"Enter {param} value: "))  # Taking user input
                values[param] = value
                break  # Exit loop if input is valid
            except ValueError:
                print(" Invalid input! Please enter a numeric value.")

    # Check if all values are within the safe range
    for param, value in values.items():
        min_val, max_val = safe_limits[param]
        if not (min_val <= value <= max_val):
            print(f"\n {param} ({value}) is OUT of safe range ({min_val} - {max_val})")
            print("\n Water is **NOT SAFE** for Drinking.\n")
            return  # Exit function if any value is unsafe

    print("\n Water is **SAFE** for Drinking.\n")

# Run the function
check_water_safety()
```
